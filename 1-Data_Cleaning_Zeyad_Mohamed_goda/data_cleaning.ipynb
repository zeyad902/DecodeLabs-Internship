```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r"C:\Users\ysyzy\.ipynb_checkpoints\decode_intern\Dataset for Data Analytics.xlsx")
```

```python
## there is 309 row blank in couponcode i will fill them with no coupon 
df['CouponCode'].fillna("No_coupon",inplace=True)
```

```python
df['CouponCode'].isnull().sum()
## there is no null values in coupon code
```

```python
df.duplicated().sum()
## there is no duplicates in the data
```

```python
discount_map = {
    'SAVE10' : 0.10,
    'WINTER15' : 0.15,
    'FREESHIP' : 0.0,
    'No_coupon' : 0.0
}

df['discount_rate'] = df['CouponCode'].map(discount_map)
```

```python
df = df.rename(columns={'TotalPrice':'gross_total'})
# I change the total price to grosstotal (price before discount)
```

```python
df['discount amount'] = (df['discount_rate'] * df['gross_total']).round(2)
df['net_total'] = (df['gross_total'] - df['discount amount']).round(2)
```

```python
df.columns = (
    df.columns
    .str.strip()           # remove accidental spaces around the name
    .str.lower()           # everything lowercase
    .str.replace(' ', '_') # spaces become underscores
)
```

```python
df = df.rename(columns={
    'orderid':          'order_id',
    'customerid':       'customer_id',
    'unitprice':        'unit_price',
    'shippingaddress':  'shipping_address',
    'paymentmethod':    'payment_method',
    'orderstatus':      'order_status',
    'trackingnumber':   'tracking_number',
    'itemsincart':      'items_in_cart',
    'couponcode':       'coupon_code',
    'referralsource':   'referral_source',
})

print(df.columns.to_list())
```

```python
df.to_csv('cleaned_file.csv',index=False)
```
